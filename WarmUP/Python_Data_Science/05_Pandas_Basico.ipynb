<span style="font-size:14px;">Vimos no último vídeo que é possível criar DataFrames e Series a partir de várias fontes (arquivos externos, listas, dicionários, etc).  
  
<span style="font-size:14px;">Na aula que falamos sobre funções, desenvolvemos em nossos exercícios uma função que recebia um dicionário com um conjunto de informações sobre veículos, e calculava a quilometragem média anual de cada veículo. Esta função retornava o conteúdo do dicionário de input da função, incluindo as informações sobre a quilometragem média.  
  
<span style="font-size:14px;">Dica: Para resolver esta questão, será necessário relembrar de um recurso que aprendemos no curso anterior, quando falamos de arrays Numpy. Este recurso também pode ser aplicado a DataFrames do pandas:  
  
<span style="font-size:14px;">ndarray.T: Retorna o array transposto, isto é, converte linhas em colunas e vice versa.

```python
dados = {
    'Crossfox': {'km': 35000, 'ano': 2005}, 
    'DS5': {'km': 17000, 'ano': 2015}, 
    'Fusca': {'km': 130000, 'ano': 1979}, 
    'Jetta': {'km': 56000, 'ano': 2011}, 
    'Passat': {'km': 62000, 'ano': 1999}
}
```

```python
def km_media(dataset, ano_atual):
    result={} # linha 3 a 6 é tudo o que eu devo ter dentro do result / resultado desta função
    for item in dataset.items(): # para cada item/linha do dataset terei:
        media = item[1]['km'] / (ano_atual - item[1]['ano']) #media de quilomentragem,
        item[1].update({'km_media': media}) #estou acrescentando esta informação no dicionário / , a informação (nova) 'km_media' e o valor da quilometragem da média (linha4)
        result.update({item[0]:item[1]}) # a informação acima será alocada no item [0] = chave e no item [1] = valores correspondentes as chaves
    return result
```

```python
km_media(dados, 2019)
```

```python
# Construindo um DataFrame, a partir do dicionario acima, onde a coluna chave seja os nomes dos carros e values/atributos sejam o que esta 
#dentro do dicionário (km, ano e km_media)
carros = pd.DataFrame(km_media(dados, 2019))
```

```python
carros
```

```python
# Transpondo o DataFrame, basta colocar o T no final do código
carros = pd.DataFrame(km_media(dados, 2019)).T
```

```python
carros
```

Queries com DataFrames  
  
... que são consultas que podem ser feitas no DataFrame  
  
Queremos realizar uma query que nos traga todos os registros para determinado tipo de motor, por exemplo o "Motor Diesel". Antes de fazermos isso, é interessante comentar que existe outra forma de obtermos uma series simples a partir do rótulo de um dataframe, que é a sintaxe dataframe.Rotulo, como no exemplo abaixo:

```python
dataset.head()
```

```python
# Selecionando uma coluna
dataset.Motor
```

Observação: Em Python, == é um operador de comparação usado para verificar se dois valores são iguais. Ele retorna True se os valores comparados forem iguais e False caso contrário.

```python
# Quero que me retorne todos os carros que são 'Motor Diesel' / Vai me retornar resultado boleano (True or False)
dataset.Motor == 'Motor Diesel'
```

```python
# Essa series pode ser passada como parâmetro em uma seleção de dataframe para obtermos somente os registros aos quais o valor "True" foi atribuído. 
# Para facilitarmos esse tipo de consulta, armazenaremos nossa series em uma variável select.
select = dataset.Motor == 'Motor Diesel'
```

```python
type(select)
```

```python
dataset[select]
```

Vamos supor que queremos realizar uma query adicional, dessa vez verificando se esses veículos também são "Zero_km". Para isso, a partir do nosso dataset, passaremos entre parênteses dois pares de parênteses separados por um & (and, caractere conhecido como "e comercial").

```python
dataset[(select) & (dataset.Zero_km==True)]
```

Existe outra forma de fazermos queries nos dataframes do Pandas, que é por meio da função query(). Com ela também podemos passar uma seleção, mas seguindo uma sintaxe um pouco diferente, na qual nossas comparações serão cercadas por aspas simples ou duplas. Note que se for necessário passar strings na seleção, elas deverão ser sinalizada pelo caractere que não foi utilizado na seleção superior - ou seja, se nossa seleção está cercada de aspas simples, a string deverá ser passada dentro de aspas duplas e vice-versa.

```python
dataset.query('Motor=="Motor Diesel" and Zero_km == True')
```

Além do "and" (&), o operador "or" (|, "ou") também pode nos auxiliar nas nossas seleções.

```python
dataset[(dataset.Motor == 'Motor Diesel') | (dataset.Zero_km == True)]
```

```python
# ou podemos fazer da seguinte forma:
dataset.query('Motor == "Motor Diesel" or Zero_km == True')
```

5.4 Iterando com DataFrames

```python
dataset.head()
```

```python
# não é a forma correta!
for item in dataset:
    print(item)
```

A forma abaixo para iterar dentro de um DataFrame serve para bases pequenas... não é ideal para bases grandes de dados

```python
# Forma correta - Desempacotar as tuplas 
for index, row in dataset.iterrows(): # itera pelas linhas do índice
    if(2019 - row['Ano'] != 0): # se a linha do ano for diferente de 0
        dataset.loc[index, 'km_media'] = row['Quilometragem'] / (2019 - row['Ano']) # localize o indice e inclua uma coluna com o nome de 'km_media' e faça o cálculo desta linha que é a quilometragem desta linha, dividido pelo ano atual (2019) - o Ano dessa linha
    else: dataset.loc[index,'km_media']=0 # se não atribua o valor = 0 / para a nova coluna 'km_media'.
dataset
```
